```python
#importing all mandatory libraries, resolving dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#displays version of seaborn
sns.__version__
```

```python
#Load sample dataset using fucntion 'load_dataset()'
tips = sns.load_dataset('tips')

sns.set()
```

```python
tips.tail(n=10)                                     #displays initial rows from included dataframe using fucntion head()
```

total_bill  : Total bill (cost of the meal), including tax, in US dollars <br/>
tip         : Tip (gratuity) in US dollars <br/>
sex         : Sex of person paying for the meal (0=male, 1=female) <br/>
smoker      : Smoker in party? (0=No, 1=Yes)<br/>
day         : 3=Thur, 4=Fri, 5=Sat, 6=Sun<br/>
time        : 0=Day, 1=Night<br/>
size        : Size of the party<br/>

```python
#faceted scatter plot with multiple variables
#scatter plot

sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            kind="scatter", data=tips, legend="full")
```

```python
sns.relplot(x="total_bill", y="tip", col="sex",
            hue="smoker", style="time", data=tips)
```

```python
#Line Plot are drawn using the fucntion relplot() with attribute kind="line"
sns.relplot(x="total_bill", y="tip", col="sex", 
            kind="line", data=tips)
```

```python
#categorical plots are displayed using function catplot()
sns.catplot(x="day", y="total_bill", col="sex", data=tips);
```

```python
#Box and whisker plot is diplayed using attribute kind="box"
sns.catplot(x="day", y="total_bill", kind="violin", data=tips);
```

```python
#Linear Regression plot, function regplot() and lmplot()
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips)
```
