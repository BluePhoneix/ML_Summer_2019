```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Scatter plot in matplotlib
d = pd.read_csv('tips.csv')
tips= d['tip']
total= d['total_bill']
plt.scatter(tips, total, label="Scatter", color="g")

plt.xlabel("tips")
plt.ylabel("Total_Bill")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

```python
#loads sample dataset
tips = sns.load_dataset('tips')

#applies default theme
sns.set()

#plotting a simple scatter plot using seaborn
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```
