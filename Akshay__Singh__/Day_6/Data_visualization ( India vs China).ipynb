```python
#Now, Let's working on Actual Data(different countries_population_data)
data = pd.read_csv('countries.csv')
```

```python
#shows data_points
data
```

```python
# Now let's compare the population growth of India and China
```

```python
# Indian population
data[data.country == 'India']
```

```python
# let's take it in variable for futher use
Ind = data[data.country == 'India']
```

```python
# Same thing for china well
china = data[data.country == 'China']
```

```python
#Chinees Population
china
```

```python
# let's compare their Population Growth
plt.plot(Ind.year, Ind.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['India', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
Ind.population
```

```python
# Let's check how growth is going on from first year by taking
Ind.population / Ind.population.iloc[0] * 100
```

```python
#finally
plt.plot(Ind.year, Ind.population / Ind.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['India', 'China'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```
