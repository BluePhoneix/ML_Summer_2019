# Using litle styling in plot by importing style from matplotlib


```python
from matplotlib import style
style.use('ggplot')
x = [5,12,15]
y = [12,16,5]
x2 = [6,9,11]
y2 = [6,15,16]

plt.plot(x,y,'g',label ='line one',linewidth=3)
plt.plot(x2,y2,'c',label ='line two',linewidth=3)
plt.title('style graph')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.legend()
plt.show
```

# Plotting  bar graph


```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="example1")
plt.bar([2,4,6,8,10],[8,6,2,5,6],label="example2")
plt.legend()
plt.xlabel("bar no")
plt.ylabel("bar height")
plt.title('bar graph')
plt.show
```
