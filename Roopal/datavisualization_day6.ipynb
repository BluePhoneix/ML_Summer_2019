```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#example of subplots, the three values under the subplot function tells number of rows column and the number of plots repectively 
plt.figure(figsize=(30,8))
#the parameters in fig to specify the lenght of the subplot
plt.subplot(131)
 #Code for fig1.
plt.subplot(132)
#code for fig2
plt.subplot(133)
#code for fig3.
plt.show()
#for better view of the subplot
plt.grid()
```

```python
print(df["SepalWidthCm"])
```

```python
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'],c=['g'])
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter plot on Iris dataset')
```

```python
#to show the different labels in different colors facet grid represent the data according to their occurance height        
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="Species", height=4)\
  .map(plt.scatter,"SepalLengthCm","SepalWidthCm" )\
 .add_legend()
```

```python
f=df.drop("Id",axis=1)
```

```python
# Pairplot these are plotted according to dimension of features
sns.set_style("whitegrid");
sns.pairplot(f,hue="Species",height=4)
plt.show()
```

```python
#for regression purpose we assume a line along which the data varies implot gives that line of regression
sns.lmplot(x="SepalLengthCm", y="SepalWidthCm", data=df, hue="Species")
```

```python
#boxplot is univariate and represents the quartiles and the whiskers
#here is a demonstration on how the class species varies according to sepal width
sns.boxplot(x='Species',y='SepalWidthCm',data=df)
```

```python
sns.violinplot(x='Species',y='SepalWidthCm',data=f,size=6)
```

```python
#multivariate graph
sns.FacetGrid(f, hue= 'Species', size= 5)\
   .map(sns.distplot, "PetalWidthCm")\
   .add_legend()
plt.show()
```
